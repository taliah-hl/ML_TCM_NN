# Just to try NN

very simple example to show how to use sequential model in tensorflow.keras

using self-generated data with 1 feature

0=no side effects; 1= has side effects

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels =[]
train_samples=[]
```

### create dummy data


Example data:

• An experiemental drug was tested on individuals from ages 13 to 100 in a clinical trial.

• The trial had 2100 participants. Half were under 65 years old, half were 65 years or older. 

• Around 95% of patients 65 or older experienced side effects. (1)

• Around 95% of patients under 65 experienced no side effects. (0)

• 0=no side effects; 1= has side effects

```python
for i in range(50):
    # The ~5% of younger individuals who did experience side e 
    random_younger = randint(13,64) 
    train_samples.append(random_younger)
    train_labels.append(1)

    # The ~5% of older individuals who did not experience side effects 
    random_older = randint(65,100) 
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    # The ~95% of younger individuals who did not experience side effects 
    random_younger = randint(13,64) 
    train_samples.append(random_younger)
    train_labels.append(0)

    # The ~95% of older individuals who did experience side effects 
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)

# my note:
# x, y is in same dimension
```

```python
print("len of train_samples:" ,len(train_samples))
print("len of train_labels:" ,len(train_labels))
```

```python
# randomized the generated data

train_labels = np.array(train_labels)
train_samples = np.array(train_samples)

train_labels, train_samples = shuffle(train_labels, train_samples)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
# .reshape(-1,1) in this case change 1D array to 2D column array (row vector to col vector)
# MinMaxScaler and fit_transform together transform age range to 0 to 1
```

```python
for i in scaled_train_samples:
    print(i)
```

```python
print()
```

Build Simple tf.keras Sequential Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
# run this only if need GPU
# physical_devices = tf.config.experimental.list_physical_devices ('GPU') 
# print("Num GPUs Available: ", len(physical_devices)) 
# tf.config.experimental.set_memory_growth (physical_devices[0], True)
```

```python
model = Sequential([
    Dense (units=16, input_shape=(1,), activation='relu'),
    Dense (units=32, activation='relu'), 
    Dense (units=2, activation='softmax')
])

# Sequential(<list of hidden layers>)
# first hidden layer need to specify input dimension in input_shape
# input_shape=(1,) -> input is n*1 array i.e. 1 feature
# if input is n*2 array (2 features) -> input_shape=(2,) 
# last Dense is output layer, units must = no. of possible output
# softmax give probi of output
```

```python
model.summary()
```

Train model

```python
# compile the traing
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# metrics specify what to use to judge the model in training process
# commonly just use accuracy
```

```python
# actually train the model
# no validation version

model.fit(x=scaled_train_samples, y=train_labels,
          batch_size=10, epochs=30, shuffle=True, verbose=2)

# batch_size: how large 1 1 batch passed to network each time (????)
# epochs = train多少次
# shuffle by default is true 防止model 從次序learn
# verbose: level of verbose message to show (0/1/2)(larger=more detail msg)
```

```python
# training with validation version
# note: model 會分出validation set first before shuffle
# e.g. validation_split = 0.1 means last 10% of data before shuffle is used as validation
# so need to shuffle before calling model.fit!

model.fit(x=scaled_train_samples, y=train_labels,
          validation_split = 0.1,
          batch_size=10, epochs=30, shuffle=True, verbose=2)
```

Prediction

```python
predictions = model.predict(x=scaled_train_samples, batch_size=10, verbose=0)
# 不清楚batch size是否要和train時一樣
```

```python
# predictions[i][0] = probi the 0th possible outupt
# predictions[i][1] = probi the 1st possible outupt etc.
#
# e.g. in this case  0=no side effects; 1= has side effects

print((type(predictions[0])))
print((type(predictions[0][0])))
for i in range(10):
    print(predictions[i])
```

```python
# extract the output with highest probi (most probable outcome)
rounded_predictions = np.argmax(predictions, axis=-1) # axis=-1 along row
```
